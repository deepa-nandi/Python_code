```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
import numpy as np


def get_title(soup):
    try:
        title = soup.find('span', attrs={'id':'productTitle'})
        title_string=title.text.strip()

    except Error:
        title_string=" "

    return title_string

def get_price(soup):
    try:
        price = soup.find('span', attrs={'class':'a-size-medium a-color-price'})
        price_string = price.text.strip()

    except:
        price_string = " "

    return price_string

def get_rating(soup):
    try:
        rating = soup.find('span', attrs={'class':'a-size-base a-color-base'})
        rating_string = rating.text.strip()

    except:
        rating_string = " "
    
    return rating_string

def get_review(soup):
    try:
        review = soup.find('span', attrs={'id':'acrCustomerReviewText'})
        review_string = review.text.strip()

    except:
        review_string = " "

    return review_string


if __name__ == '__main__':

    HEADERS = ({'User-Agent':'Mozilla/5.0 (Macintosh; Intel Mac OS X 10.15; rv:109.0) Gecko/20100101 Firefox/113.0', 'Accept-Language':'en-US, en;q=0.5'})

    URL = "https://www.amazon.com/s?k=playstation+4&crid=FYI8JDVZ4F5U&sprefix=playstation+%2Caps%2C323&ref=nb_sb_noss_2"

    webpage = requests.get(URL, headers=HEADERS)

    soup = BeautifulSoup(webpage.content,"html.parser")

    links = soup.find_all("a", attrs={'class':'a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal'})

    link_list = []

    for i in links:
        link_list.append(i.get('href'))

    d = {"title":[], "price":[], "ratings":[], "reviews":[]}

    for i in link_list:
        new_webpage = requests.get("https://amazon.com" + i, headers=HEADERS)

        new_soup = BeautifulSoup(new_webpage.content,"html.parser")

        d['title'].append(get_title(new_soup))
        d['price'].append(get_price(new_soup))
        d['ratings'].append(get_rating(new_soup))
        d['reviews'].append(get_review(new_soup))


    amazon_df = pd.DataFrame.from_dict(d)
    amazon_df['title'].replace('',np.nan, inplace=True)
    amazon_df = amazon_df.dropna(subset=['title'])
    amazon_df.to_csv("amazon_data.csv", header=True, index=False)
```
